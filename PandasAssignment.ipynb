# Q11. How do you group data in a Pandas DataFrame by a specific column and apply an aggregation function?

```python
df = pd.DataFrame({
    'name' : ['Ajay', 'Ananya', 'Babu','Bhavani', 'Chethana'],
    'gender' : ['M','F','M','F','F'],
    'age' : [33,22,33,22,12]
})

group = df.groupby(by=['gender'])
group.groups
```

```python
#Aggreation
group.agg('sum')
```

# Q19. How do you calculate the correlation between two columns in a Pandas DataFrame?

# Q25. What is the difference between .loc and .iloc in Pandas?

```python
# .loc - refers to location. loc[] is used to select rows and columns by Names/Labels
# .iloc - referes to Integer location/position. 

df = pd.DataFrame({
    'Maths' : [78,85,67,69],
    'Physics' : [81,77,63,74],
    'History' : [53,65,95,87]
})

df
```

```python
# Access elements using iloc  - Integer positions
df.iloc[0:2,0:1]
```

```python
# Accessing elements using loc - Labels/Names
df.loc[0:2, "History"]
```
